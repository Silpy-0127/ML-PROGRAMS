```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import export_text
import matplotlib.pyplot as plt
```

```python
x = np.array([[150,0],[170,1],[120,0],[140,1],[200,1],[130,0]])
y = np.array(['apple','orange','apple','orange','melon','apple'])
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(x,y)
```

```python
tree_rules = export_text(clf, feature_names=['weight','texture'], class_names=np.unique(y))
print("decison tree calssifier rules\n", tree_rules)
```

```python
plt.figure(figsize=(10,6))
plot_tree(clf, filled=True, feature_names=['Weight','Texture'], class_names=np.unique(y))
plt.show()
```
