```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x = np.array([[30,50000],[35,60000],[40,80000],[25,30000],[45,100000],[20,20000],[50,120000],[55,150000],[60,140000],[28,40000]])
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(x)
```

```python
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(8,6))
#plt.scatter(x[:, 0], centers[:, 1], ...) # The original line caused the error.
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis') # Scatter plot of data points, colored by cluster labels.
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('k-Means Clustering of Csutomers')
plt.legend()
plt.show()
```
